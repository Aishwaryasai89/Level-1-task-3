```python
import pandas as pd
import numpy as np
```

```python
pd.set_option("display.max_columns",21)
```

```python
df = pd.read_csv('Dataset .csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (8,6))
df["Price range"].value_counts().sort_index().plot(kind="bar")
plt.xlabel("Price range")
plt.ylabel("Rumber of Restaurants")
plt.title("Distribution of price range among Restaurants")
plt.xticks(rotation = 0)
plt.show()
```

```python
total_restaurants = df["Restaurant Name"].count()
```

```python
price_range_counts = df["Price range"].value_counts()
```

```python
percentage_restaurants = (price_range_counts/total_restaurants)*100
print(f"Percentage of restaurants in each price range category:\n{percentage_restaurants}%")
```
